## Creating html file list

```python
import pandas as pd
```

```python
import os
file_list = [f for f in os.listdir('htmls') if f.endswith(".html")]
len(file_list)
print(file_list)
```

```python
data_companies=dict()
for i in range(len(file_list)):
    a= pd.read_html('htmls/'+str(file_list[i]))
    data_companies[(file_list[i]).split('.html')[0]] = a
```

## from nested dataframes in dictionary to a dataframe

```python
citility=pd.concat([pd.DataFrame(data_companies['citility'][0]), pd.DataFrame(data_companies['citility'][1]), pd.DataFrame(data_companies['citility'][2][0:3]), pd.DataFrame(data_companies['citility'][4])], axis=0).reset_index(drop=True)
```

```python
new_df=pd.concat([pd.DataFrame(data_companies['citility'][0][0]), pd.DataFrame(data_companies['citility'][1][0]), pd.DataFrame(data_companies['citility'][2][0][0:3])], axis=0).reset_index(drop=True)
new_df.columns=['ind']
for i in list(data_companies.keys()):
    a= pd.concat([pd.DataFrame(data_companies[str(i)][0]), pd.DataFrame(data_companies[str(i)][1]), pd.DataFrame(data_companies[str(i)][2][0:3])], axis=0).reset_index(drop=True)   
    a.columns=['ind', i]
    new_df= new_df.merge(a, how='outer')
```

```python
new_df
```

```python
(new_df.set_index("ind").T).to_csv('companies_info_data.csv', sep=';')
```

```python
finance_df=pd.DataFrame(data_companies['citility'][3])
for i in list(data_companies.keys()):
    if len(data_companies[str(i)][3]) > 10:
        b= pd.DataFrame(data_companies[str(i)][3]).set_index("Performance").add_suffix('_'+i)
        finance_df=finance_df.merge(b, how='outer', on="Performance")
finance_df.drop(columns=['2017', '2016'], inplace=True)
```

```python
(finance_df.set_index("Performance").T).to_csv("companies_finance_data.csv", sep=';')
```
